2. Crawl CNN with related articles to climate change. In this section, I call the search website API and render each datapoint into csv format to parse each site later

```python
upperframe=[]
def scarp(url):
  website=requests.get(url)
  soup=BeautifulSoup(website.content, 'lxml')
  news_content=''
  for news in soup.findAll('div', {'data-editable':'content'}):
    news_content+=news.text.strip()

def main(url):

    with requests.Session() as req:
        filename="New.CSV"

        for item in range(1, 100):
            r = req.get(url.format(item)).json()
            frame=[]
            f=open(filename,"w", encoding = 'utf-8')
            headers="Headline,Link,Date, Brief"
            f.write(headers)
            for a in r['result']:
                Headline=a['headline']
                Link=a['url']
                Date=a['lastModifiedDate']
                Brief=a['body']
                #Content = scarp(a['url'])
                #print("Headline: {}, Url: {}, Date:{}".format(
                #    a['headline'], a['url'], a['lastModifiedDate']))
                frame.append((Headline,Link,Date,Brief))
                f.write(Headline.replace(",","^")+","+Link+","+Date.replace(",","^")+","+Brief.replace(",","^"))
    upperframe.extend(frame)
    f.close()



main("https://search.prod.di.api.cnn.io/content?q=climate%20change&size=100&from=0&page=1&sort=newest")
data=pd.DataFrame(upperframe, columns=['Headline','Link','Date','Brief'])
data.head(30)
```

```python
#load regular expresion library
import re

# Remove the columns
news = combined.drop(columns=['Link', 'Date', 'Brief'], axis=1)

combined
# Remove punctuation
news['processed'] = \
news['Content'].map(lambda x: re.sub('[,\.!?]', '', x))
news['Content'].map(lambda x: re.sub('\xa0', '', x))

# Convert the titles to lowercase
news['processed'] = \
news['processed'].map(lambda x: x.lower())

# Print out the first rows of papers
news['processed'][0]
```

```python
print(news.dtypes)
```

```python
#Tokenization and delete stop words

#Tokenization
def tokenize(text):
    split=re.split("\W+",text)
    return split
news['token']=news['processed'].apply(lambda x: tokenize(x.lower()))
news.head()
```

```python
#Delete stop words I you we they etc
import nltk
nltk.download('stopwords')
stopword = nltk.corpus.stopwords.words('english')
print(stopword[:11])

def remove_stopwords(text):
    text=[word for word in text if word not in stopword]
    return text
news['token'] = news['token'].apply(lambda x: remove_stopwords(x))
news.head()
```

# C. Exploratory Analysis

```python
# Import the wordcloud library
from wordcloud import WordCloud

# Join the different processed titles together.
long_string = ','.join(list(news['processed'].values))

# Create a WordCloud object
wordcloud = WordCloud(background_color="white", max_words=1000, contour_width=3, contour_color='steelblue')

# Generate a word cloud
wordcloud.generate(long_string)

# Visualize the word cloud
wordcloud.to_image()
```

```python
#delete words that are not necessary
delword=['said', 'getty', 'image','cnn', 'according', 'told']
def remove_word(text):
  text=[word for word in text if word not in delword]
  return text
news['token'] = news['token'].apply(lambda x: remove_word(x))

#create world cloud again
news['token_str'] = news['token'].apply(lambda x: ' '.join(x))
long_string2 = ' '.join(news['token_str'])
wordcloud2 = WordCloud(background_color="white", max_words=1000, contour_width=3, contour_color='steelblue')
wordcloud2.generate(long_string2)
wordcloud2.to_image()
```

```python
#Lemmatize/ Stem: - lemmatizing take context into consideration
#reducing a word to its root form

from nltk.stem import WordNetLemmatizer

# Download the WordNet lemmatizer data (only required for the first time)
nltk.download('wordnet')

# Create an instance of the WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

news['lemmatized_tokens'] = news['token'].apply(lambda tokens: [lemmatizer.lemmatize(word) for word in tokens])
```

# D. Machine Learning Models

## 1. Topic Modeling - LDA model

```python
#Next, we convert the tokenized object into a corpus and dictionary.

import gensim.corpora as corpora

# Create Dictionary
id2word = corpora.Dictionary(news['lemmatized_tokens'])

# Create Corpus
texts = news['lemmatized_tokens']

# Term Document Frequency
corpus = [id2word.doc2bow(text) for text in texts]

# View
print(corpus[:1][0][:30])
```

```python
from pprint import pprint

# number of topics
num_topics = 10

# Build LDA model
lda_model = gensim.models.LdaMulticore(corpus=corpus,
                                       id2word=id2word,
                                       num_topics=num_topics,
                                       passes=1)

# Print the Keyword in the 10 topics
pprint(lda_model.print_topics())
doc_lda = lda_model[corpus]
```
